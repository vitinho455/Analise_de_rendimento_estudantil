```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

table = pd.read_csv('student_habits_performance.csv')

study_hours = ['study_hours_per_day', 'social_media_hours']

for coluna in study_hours:

    plt.figure(figsize=(6, 4))
    sns.histplot(table[coluna], kde=True)
    plt.title("Histogram")
    plt.xlabel(coluna)
    plt.ylabel("Frequency")
    plt.show()
```
